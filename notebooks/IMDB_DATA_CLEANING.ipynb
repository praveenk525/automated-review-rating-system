REMOVING THE UNNECESSARY COLUMNS

```python
df=df.drop(columns=['Unnamed: 0', 'Show ID', 'Review ID'])
```

CHECKING NULL VALUES

```python
df.isnull().sum()
```

RENAMING THE COLUMNS

```python
df.rename(columns={'Rating (out of 10)':'rating'},inplace=True)
```

CHECKING DUPLICATES

```python
print(df.duplicated().sum())
```

DROPPING DUPLICATES

```python
df.drop_duplicates(subset=['Review','rating'],inplace=True)
```

CONVERT RATING SCALE(10 TO 5)

```python
df['rating']=df['rating']/2
```

```python
df['rating']=df['rating'].astype(int)
```

REMOVING EMOJIS,LINKS

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "Review"] = df["Review"].apply(clean_text)
```

```python
df['rating'].value_counts()
```

REMOVING 0 RATING

```python
df = df[df['rating'] != 0]
```

SORTING THE DATA

```python
df.sort_values(by='rating',ascending=True,inplace=True)
```

SAVING THE DATASET

```python
df.to_csv('clean_imdb.csv',index=False)
```
