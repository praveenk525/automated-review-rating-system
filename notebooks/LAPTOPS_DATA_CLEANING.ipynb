REMOVING THE UNNECESSARY COLUMNS

```python
df=df.drop(columns=['product_name', 'overall_rating', 'no_ratings', 'no_reviews',
       'title'])
```

CHECKING NULL VALUES

```python
df.isnull().sum()
```

CHECKING DUPLICATES

```python
print(df.duplicated().sum())
```

DROPPING DUPLICATES

```python
df.drop_duplicates(subset=['review','rating'],inplace=True)
```

```python
df['rating'].astype(int)
```

REMOVING EMOJIS,LINKS

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "review"] = df["review"].apply(clean_text)
```

```python
df['rating'].value_counts()
```

SORTING THE DATA

```python
df.sort_values(by='rating',ascending=True,inplace=True)
```
