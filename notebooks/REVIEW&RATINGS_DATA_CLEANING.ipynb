REMOVING THE UNNECESSARY COLUMNS

```python
df=df.drop(columns=['Unnamed: 0',  'Lang', 'Type', 'Country', 'Date', 'Helpful',])
```

RENAMING THE COLUMNS

```python
df.rename(columns={'translated':'review'},inplace=True)
```

CHECKING NULL VALUES

```python
df.isnull().sum()
```

CHECKING DUPLICATES

DROPPING DUPLICATES

```python
df.drop_duplicates(subset=['review','Rating'],inplace=True)
```

```python
df['Rating'].astype(int)
```

REMOVING EMOJIS,LINKS

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "review"] = df["review"].apply(clean_text)
```

```python
df['Rating'].value_counts()
```

SORTING THE DATA

```python
df.sort_values(by='Rating',ascending=True,inplace=True)
```
