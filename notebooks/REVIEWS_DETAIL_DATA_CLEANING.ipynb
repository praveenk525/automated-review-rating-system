REMOVING THE UNNECESSARY COLUMNS

```python
df=df.drop(columns=['report_date', 'online_store', 'upc', 'retailer_product_code', 'brand',
       'category', 'sub_category', 'product_description', 'review_date',
       'review_title','is_competitor',
       'manufacturer', 'market', 'matched_keywords', 'time_of_publication',
       'url', 'review_type', 'parent_review', 'manufacturers_response',
       'dimension1', 'dimension2', 'dimension3', 'dimension4', 'dimension5',
       'dimension6', 'dimension7', 'dimension8', 'verified_purchase',
       'helpful_review_count', 'review_hash_id'])
```

RENAMING THE COLUMNS

```python
df.rename(columns={'review_rating':'rating','review_text':'review'},inplace=True)
```

CHECKING NULL VALUES

```python
df.isnull().sum()
```

CHECKING DUPLICATES

```python
print(df.duplicated().sum())
```

DROPPING DUPLICATES

```python
df.drop_duplicates(subset=['review','rating'],inplace=True)
```

```python
df['rating'].astype(int)
```

REMOVING EMOJIS,LINKS

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "review"] = df["review"].apply(clean_text)
```
