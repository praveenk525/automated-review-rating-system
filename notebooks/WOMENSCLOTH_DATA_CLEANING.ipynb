REMOVING THE UNNECESSARY COLUMNS

```python
df=df.drop(columns=['Unnamed: 0', 'Clothing ID', 'Age', 'Title',
       'Recommended IND', 'Positive Feedback Count', 'Division Name',
       'Department Name', 'Class Name'])
```

CHECKING NULL VALUES

```python
df.isnull().sum()
```

DROPPING THE NULL VALUES

```python
df.dropna(inplace=True)
```

CHECKING DUPLICATES

```python
print(df.duplicated().sum())
```

DROPPING DUPLICATES

```python
df.drop_duplicates(subset=['Review Text','Rating'],inplace=True)
```

```python
df['Rating'].astype(int)
```

REMOVING EMOJIS,LINKS

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "Review Text"] = df["Review Text"].apply(clean_text)
```

```python
df['Rating'].value_counts()
```

SORTING THE DATA

```python
df.sort_values(by='Rating',ascending=True,inplace=True)
```

SAVING THE DATASET

```python
df.to_csv('clean_womensclothing.csv',index=False)
```
