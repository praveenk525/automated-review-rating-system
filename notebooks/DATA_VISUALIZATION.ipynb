LOADING THE DATASETS

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
cl_df=pd.read_csv('/content/cleaned_dataset.csv')
bal=pd.read_csv('/content/Balanced_dataset.csv')
```

BAR CHART(CLEANED DATASET)

```python
x=cl_df['rating'].value_counts().index
y=cl_df['rating'].value_counts().values
sns.barplot(x=x, y=y)
plt.title("BAR PLOT-RATINGS COUNT")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()
```

HORIZONTAL BAR PLOT(CLEANED DATASET)

```python
plt.barh(x,y)
plt.title("HORIZONTAL BAR PLOT-RATINGS COUNT")
plt.xlabel("Count")
plt.ylabel("Ratings")
plt.show()
```

PIE CHART(CLEANED DATASET)

```python
cl_df['rating'].value_counts().sort_index().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title("Ratings Proportion")
plt.ylabel('')
plt.show()
```

BAR CHART(BALANCED DATASET)

```python
x=bal['rating'].value_counts().index
y=bal['rating'].value_counts().values
sns.barplot(x=x, y=y)
plt.title("BAR PLOT-RATINGS COUNT")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()
```

HORIZONTAL BAR CHART(BALANCED DATASET)

```python
plt.barh(x,y)
plt.title("HORIZONTAL BAR PLOT-RATINGS COUNT")
plt.xlabel("Count")
plt.ylabel("Ratings")
plt.show()
```

```python
bal['rating'].value_counts().sort_index().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title("Ratings Proportion")
plt.ylabel('')
plt.show()
```
