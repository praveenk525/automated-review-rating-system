REMOVING THE UNNECESSARY COLUMNS

```python
df=df.drop(columns=['Unnamed: 0', 'recipe_number', 'recipe_code', 'recipe_name',
       'comment_id', 'user_id', 'user_name', 'user_reputation', 'created_at',
       'reply_count', 'thumbs_up', 'thumbs_down',  'best_score',
       ])
```

CHECKING NULL VALUES

```python
df.isnull().sum()
```

DROPPING THE NULL VALUES

```python
df.dropna(inplace=True)
```

RENAMING THE COLUMNS

```python
df.rename(columns={'text':'review','stars':'rating'},inplace=True)
```

CHECKING DUPLICATES

```python
print(df.duplicated().sum())
```

DROPPING DUPLICATES

```python
df.drop_duplicates(subset=['review','rating'],inplace=True)
```

```python
df['rating'].astype(int)
```

REMOVING EMOJIS,LINKS

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "review"] = df["review"].apply(clean_text)
```

```python
df['rating'].value_counts()
```

REMOVING 0 RATING

```python
df = df[df['rating'] != 0]
```

SORTING THE DATA

```python
df.sort_values(by='rating',ascending=True,inplace=True)
```

SAVING THE DATASET

```python
df.to_csv('clean_recipe.csv',index=False)
```
